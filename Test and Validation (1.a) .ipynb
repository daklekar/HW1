```python
train = train[['price','area','bedrooms','bathrooms','stories','parking']]
test = test[['price','area','bedrooms','bathrooms','stories','parking']]
```

```python
#scaler = sklearn.preprocessing.MinMaxScaler()
```

```python
X1_t = np.array(train.area)
X1_t
```

```python
X2_t = np.array(train.bedrooms)
X3_t = np.array(train.bathrooms)
X4_t = np.array(train.stories)
X5_t = np.array(train.parking)
X0_t= np.ones(436)

```

```python
#For test set
X = np.vstack([X0_t,X1_t,X2_t,X3_t,X4_t,X5_t])
X = X.T
X = np.array(X)   
X
```

```python
#x = scaler.fit_transform(X)
#X = x
```

```python
theta = np.array([0., 0., 0., 0., 0., 0.])
theta = theta.reshape(6,1)
theta
```

```python
#h = np.matmul(X,theta)
Y_t = np.array(train.price)
Y = Y_t
Y = Y_t.reshape(436,1)
#y = scaler.fit_transform(Y)
#Y=y
```

```python
X_T = np.array(X.T)
m,n = X.shape
print(m,n)
iterations = 25

count=0
j = np.zeros(shape=(iterations, 1), dtype=float)
j.shape
```

```python
while(count < iterations):

    h_t = X.dot(theta)
    h = np.array(h_t,float)

    j[count]= (1/(2*m))*np.sum((h-Y)**2)

    grad_t = X_T.dot(h-Y)
    grad = grad_t*(1/m)

    theta = theta - 0.1*(grad)

    count = count + 1


plt.plot(j,'b-')
```

```python
V_X1_t = np.array(test.area)
V_X2_t = np.array(test.bedrooms) 
V_X3_t = np.array(test.bathrooms)
V_X4_t = np.array(test.stories)
V_X5_t = np.array(test.parking)
V_X0_t = np.ones(109)
```

```python
V_theta = np.array([0.,0.,0.,0.,0,0.])
V_theta = V_theta.reshape(6,1)
V_theta
```

```python
# V_Y = np.array(test.price)
# V_Y.reshape(109,1)
# V_y= scaler.fit_transform([V_Y])
# V_y
V_Y_t = np.array(test.price)
V_Y = V_Y_t
V_Y = V_Y_t.reshape(109,1)
#V_y = scaler.fit_transform(V_Y)
#V_Y=V_y
V_Y.shape
```

```python
V_X_T = np.array(V_X.T)
m,n = V_X.shape
m,n
```

```python
iterations = 250
count=0
V_j = np.zeros(shape=(iterations, 1), dtype=float)

while(count < iterations):

    V_h_t = V_X.dot(V_theta)
    V_h = np.array(V_h_t)


    V_j[count]= (1/(2*m))*np.sum((V_h - V_Y)**2)

    grad_t = V_X_T.dot(V_h-V_Y)
    grad = grad_t*(1/m)

    V_theta = V_theta - 0.01*(grad)

    count += 1
```
